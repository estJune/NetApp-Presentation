```python
# Assignment Operator
x = 10
y = 3

# Arithmetic Operators
print("Add:", x + y)
print("Subtract:", x - y)
print("Multiply:", x * y)
print("Divide:", x / y)
print("Modulus:", x % y)

# Comparison Operators
print("Equal:", x == y)
print("Not Equal:", x != y)
print("Greater:", x > y)
print("Less or Equal:", x <= y)

# If Statements
trending = True
views = 5000000

if trending:
    print("This video is trending!")
elif views > 1000000:
    print("This video is popular!")
else:
    print("Keep trying!")

# For Loops
for i in range(5):
    print("Number: ", i)

# While Loops
count = 0
while count < 3:
    print("Counting:", count)
    count += 1

# Lists
categories = ["Music", "Gaming", "News"]

for category in categories:
    print("Category:", category)

# Logical Operators
is_music = categories[0] == "Music"
is_gaming = categories[1] == "Gaming"
print("Music and Gaming:", is_music and is_gaming)
print("Music or Gaming:", is_music or is_gaming)


# Functions
def greet(channel):
    print("Hello,", channel)

greet("Tech Talk")
```

```python
import kagglehub
from kagglehub import KaggleDatasetAdapter
import pandas as pd

# Set the path to the file you'd like to load
file_path = "pokemon_data.csv"

# Load the latest version
df = kagglehub.load_dataset(
  KaggleDatasetAdapter.PANDAS,
  "tanishksharma9905/pokemon-data-csv",
  file_path,
)

# Convert comma separated string in the columns to lists
df["stats"] = pd.Series(df["stats"].str.split(","))
# Return dictionary for key=val being stat=val
def parse_series(series):
    return {kv.split('=')[0].strip(): kv.split('=')[1] for kv in series}
# Apply to the 'stats' column
expanded = df['stats'].apply(parse_series).apply(pd.Series)
# Convert values to numeric (optional but useful)
expanded = expanded.apply(pd.to_numeric, errors='coerce')
# Join with original DataFrame, dropping the original 'stats' column
df = pd.concat([df.drop(columns='stats'), expanded], axis=1)
# Drop the abilities column and the nan rows
df = df.drop("abilities", axis=1)
df = df.dropna()

df
```
